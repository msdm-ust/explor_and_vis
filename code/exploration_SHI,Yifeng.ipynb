## Answers of the Questions.

```python
#question 1
num=len(source_data[source_data["CURRENCY"]=="GBP"])
print("{} transactions are in GBP.".format(num))
```

```python
#question 2
num=len(source_data[source_data["CURRENCY"]!="USD"])
print("{} transactions are not in USD.".format(num))
```

```python
#question 3
pick_USD=source_data[source_data["CURRENCY"]=="USD"]
average=np.mean(pick_USD["AMOUNT"])
median=np.median(pick_USD["AMOUNT"])
print("The average and mediam transaction in USD are {} and {}, respectively.".format(average,median))
```
