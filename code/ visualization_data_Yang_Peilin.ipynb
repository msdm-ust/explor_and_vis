```python
# Usual libraries are imported here
import os
import yaml
import dask.dataframe as dd
import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv(r'C:\Users\59209\Desktop\date\catch_22_explor_and_vis-main\catch_22_explor_and_vis-main\data\german_credit_data.csv')
```

Age distribution

```python
age=np.array(data['Age'])
age_bar=np.bincount(age)
x=np.linspace(0,len(age_bar),len(age_bar))
plt.bar(x[17:],age_bar[17:])
plt.plot([np.mean(age)for i in range(50)],[i for i in range(50)],'r-')
plt.text(36,42,'mean of age')
plt.title('age distribution')
print('mean of age', np.mean(age))
```

```python
credit=np.array(data['Credit amount'])
credit_bar=np.bincount(credit)
plt.hist(credit)
plt.title('credit amount distribution')
print('mean credit amount',np.mean(credit))
```

```python
duration=np.array(data['Duration'])
plt.hist(duration)
plt.title('duration distribution')
print('mean of duration',np.mean(duration))
```

```python
purpose=data['Purpose'].groupby(data['Purpose']).count()
label=purpose.keys()
element=np.array(purpose)
plt.pie(x=element,labels=label,autopct='%3.f%%')
plt.show()
print( 'major three purpose of loan:\n car,\n radio/TV,\n furniture/equipment')
```

```python
Sex=data['Sex'].groupby(data['Sex']).count()
label=Sex.keys()
element=np.array(Sex)
plt.pie(x=element,labels=label,autopct='%3.f%%')
plt.show()
```
